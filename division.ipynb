```python
import pandas as pd #data loading, manipulation and saving
from sklearn.model_selection import train_test_split #dataset division
from sklearn.utils import shuffle # shuffling the train data
```

```python
data = pd.read_csv('thesisdata_cleaned_fullsample1.csv')
data = data.drop(columns = ['Unnamed: 0', 'url'])
data.info()
```

```python
#divide into train, validation and test and save it to data folder
training, test_full = train_test_split(data, test_size = 0.2, random_state = 123)

#internal division of train set, into purely train and validation
training_full, validation_full = train_test_split(training, test_size = 0.25, random_state = 123) # 25% here to make sure train = 60%, val = test = 20 


print(
    f"Length of train:      {training_full.shape[0]}\n"
    f"Length of validation: {validation_full.shape[0]}\n"
    f"Length of test:       {test_full.shape[0]}"
)


# shuffle the train data
training_full = shuffle(training_full, random_state = 123)

#save the files
training_full.to_csv("data_divided/training_full.csv", index = False)
validation_full.to_csv("data_divided/validation_full.csv", index = False)
test_full.to_csv("data_divided/test_full.csv", index = False)


print("the files were saved successfully")
```

```python
# remowal of outliers from the original dataset and division into train, test and validation
#initial split was already performed

# calculate Interquartile range on train only 
Q1 = training['price_numeric'].quantile(0.25)
Q3 = training['price_numeric'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

#remowing the outliers from both train and test
training_cleaned = training[(training['price_numeric'] >= lower) & (training['price_numeric'] <= upper)]
test_cleaned = test_full[(test_full['price_numeric'] >= lower) & (test_full['price_numeric'] <= upper)]

#add validation, trying to preserve the proportion
training_cleaned, validation_cleaned = train_test_split(training_cleaned, test_size = 0.25, random_state = 123)


#check the sizes
print(
    f"Length of train:      {training_cleaned.shape}\n"
    f"Length of validation: {validation_cleaned.shape}\n"
    f"Length of test:       {test_cleaned.shape}"
)

#shuffle the training set
training_cleaned = shuffle(training_cleaned, random_state = 123)


#save the files
training_cleaned.to_csv("data_divided/training_cleaned.csv", index = False)
validation_cleaned.to_csv("data_divided/validation_cleaned.csv", index = False)
test_cleaned.to_csv("data_divided/test_cleaned.csv", index = False)


print("the files were saved successfully")
```
